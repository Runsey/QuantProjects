```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Single ticker = flat DataFrame (no MultiIndex)
data = yf.download("AAPL", start="2015-01-01", end="2025-07-28")
print(type(data.columns))
price=data.xs('AAPL', axis=1, level=1)['Close']
```

```python
SMA_20 = price.rolling(window=20).mean()

SMA_50 = price.rolling(window=50).mean()
signal=SMA_50<SMA_20
signal
```

```python


plt.figure(figsize=(12, 6))
plt.plot(data['Close'], label='Close Price', color='black')
plt.plot(data['SMA_20'], label='SMA 20', linestyle='--', color='blue')

# Shade regions where signal is active (Position == 1)
plt.fill_between(data.index, data['Close'], where=data['Position'] == 1,
                 color='green', alpha=0.2, label='Signal Active (Buy)')

plt.title("AAPL Price with SMA Signal")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
daily_returns=price.pct_change()
daily_returns
```

```python
initial_price = data['Close'].iloc[0]
shares_bought = 10000/initial_price   # or initial_capital / initial_price
buy_hold = data['Close'] * shares_bought
```

```python
strategy_returns = daily_returns * signal.shift(1)
equity_curve = (1 + strategy_returns).cumprod() * 10000

plt.figure(figsize=(12, 6))
plt.plot(buy_hold , label='Buy & Hold')
plt.plot(equity_curve, label='SMA Strategy')
plt.legend()
plt.grid(True)
plt.show()
```
