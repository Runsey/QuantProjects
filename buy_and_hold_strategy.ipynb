```python
import yfinance as yf
import matplotlib.pyplot as plt

```

```python
tickers=["AAPL", "SPY"]
data= yf.download(tickers, start="2010-01-01", end="2025-07-26", auto_adjust=False)
print(data.head())  # or display(data) in Jupyter
print(data.columns)  # should include 'Adj Close'

```

```python

adj_close = data.xs('Adj Close', axis=1, level=0)


plt.figure(figsize=(12, 6))
for ticker in tickers:
    plt.plot(adj_close[ticker], label=f"{ticker} Adj Close")

plt.title("Adjusted Close Prices (20152024)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
starting_value=10000
appl_data=data.xs('AAPL',axis=1, level=1)
intial_price= appl_data["Close"].iloc[0]
shares=starting_value/intial_price
equity=shares*appl_data['Close']

plt.figure(figsize=(12, 6))
plt.plot(equity, label='AAPL Equity Curve')
plt.title("Equity Curve (Buy-and-Hold AAPL)")
plt.xlabel("Date")
plt.ylabel("Portfolio Value (USD)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
returns=appl_data['Close'].pct_change()
plt.figure(figsize=(12,6))
plt.plot(returns, label="Returns")
plt.title("Returns (Buy-and-Hold AAPL)")
plt.xlabel("Date")
plt.ylabel("Portfolio Value (USD)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
cumlative_returns=(1+returns).cumprod()-1
plt.figure(figsize=(12, 6))
plt.plot(cumlative_returns, label="AAPL Cumulative Return")
plt.title("AAPL Cumulative Return")
plt.xlabel("Date")
plt.ylabel("Return")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Assume equity is a Series of portfolio value
start_value = equity.iloc[0]
end_value = equity.iloc[-1]
n_years = (equity.index[-1] - equity.index[0]).days / 365.25

cagr = (end_value / start_value) ** (1 / n_years) - 1
print(f"CAGR: {cagr:.2%}")
```

```python
#Maximum drawdown, risk metric measureing largest percent drop from a peak to a torugh, 
rolling_max=equity.cummax()
drawdown=(equity-rolling_max)/rolling_max
MaxDrawdown=min(drawdown)
MaxDrawdown
```
